```python
!pip install pillow

from PIL import Image, ImageDraw
from IPython.display import display

# ============================================================
# LED PIXEL SIMULATION SETTINGS
# ============================================================

PIXEL_SIZE = 2      # size of each LED pixel
PIXEL_GAP = 1       # spacing between pixels

def draw_led_pixel(draw, x, y, color):
    """Draw one simulated LED pixel with spacing."""
    px = x * (PIXEL_SIZE + PIXEL_GAP)
    py = y * (PIXEL_SIZE + PIXEL_GAP)
    draw.rectangle(
        (px, py, px + PIXEL_SIZE - 1, py + PIXEL_SIZE - 1),
        fill=color
    )

# ============================================================
# 5×7 FONT (FULL UPPER + LOWERCASE + DIGITS + SYMBOLS)
# ============================================================

FONT_5x7 = {
    # Uppercase
    "A":[0x1E,0x05,0x05,0x1E,0x00],
    "B":[0x1F,0x15,0x15,0x0A,0x00],
    "C":[0x0E,0x11,0x11,0x0A,0x00],
    "D":[0x1F,0x11,0x11,0x0E,0x00],
    "E":[0x1F,0x15,0x15,0x11,0x00],
    "F":[0x1F,0x05,0x05,0x01,0x00],
    "G":[0x0E,0x11,0x15,0x1D,0x00],
    "H":[0x1F,0x04,0x04,0x1F,0x00],
    "I":[0x11,0x1F,0x11,0x00,0x00],
    "J":[0x08,0x10,0x10,0x0F,0x00],
    "K":[0x1F,0x04,0x0A,0x11,0x00],
    "L":[0x1F,0x10,0x10,0x10,0x00],
    "M":[0x1F,0x02,0x04,0x02,0x1F],
    "N":[0x1F,0x02,0x04,0x1F,0x00],
    "O":[0x0E,0x11,0x11,0x0E,0x00],
    "P":[0x1F,0x05,0x05,0x02,0x00],
    "Q":[0x0E,0x11,0x19,0x1E,0x00],
    "R":[0x1F,0x05,0x0D,0x12,0x00],
    "S":[0x12,0x15,0x15,0x09,0x00],
    "T":[0x01,0x1F,0x01,0x00,0x00],
    "U":[0x0F,0x10,0x10,0x0F,0x00],
    "V":[0x07,0x08,0x10,0x08,0x07],
    "W":[0x1F,0x08,0x04,0x08,0x1F],
    "X":[0x1B,0x04,0x04,0x1B,0x00],
    "Y":[0x03,0x04,0x18,0x04,0x03],
    "Z":[0x19,0x15,0x13,0x11,0x00],

    # Lowercase
    "a":[0x08,0x14,0x14,0x1C,0x00],
    "b":[0x1F,0x14,0x14,0x08,0x00],
    "c":[0x08,0x14,0x14,0x00,0x00],
    "d":[0x08,0x14,0x14,0x1F,0x00],
    "e":[0x0C,0x1A,0x16,0x14,0x00],
    "f":[0x04,0x1E,0x05,0x01,0x00],
    "g":[0x0C,0x12,0x12,0x1E,0x00],
    "h":[0x1F,0x04,0x04,0x18,0x00],
    "i":[0x00,0x1D,0x00,0x00,0x00],
    "j":[0x10,0x20,0x1D,0x00,0x00],
    "k":[0x1F,0x08,0x14,0x02,0x00],
    "l":[0x11,0x1F,0x10,0x00,0x00],
    "m":[0x1C,0x04,0x18,0x04,0x18],
    "n":[0x1C,0x04,0x04,0x18,0x00],
    "o":[0x08,0x14,0x14,0x08,0x00],
    "p":[0x3C,0x14,0x14,0x08,0x00],
    "q":[0x08,0x14,0x14,0x3C,0x00],
    "r":[0x1C,0x04,0x04,0x00,0x00],
    "s":[0x14,0x16,0x1A,0x0A,0x00],
    "t":[0x04,0x1F,0x14,0x00,0x00],
    "u":[0x0C,0x10,0x10,0x1C,0x00],
    "v":[0x0C,0x10,0x10,0x0C,0x00],
    "w":[0x1C,0x10,0x08,0x10,0x1C],
    "x":[0x14,0x08,0x08,0x14,0x00],
    "y":[0x0C,0x10,0x10,0x1C,0x00],
    "z":[0x14,0x1C,0x14,0x00,0x00],

    # Digits
    "0":[0x0E,0x11,0x11,0x0E,0x00],
    "1":[0x00,0x12,0x1F,0x10,0x00],
    "2":[0x19,0x15,0x15,0x12,0x00],
    "3":[0x11,0x15,0x15,0x0A,0x00],
    "4":[0x07,0x04,0x04,0x1F,0x00],
    "5":[0x17,0x15,0x15,0x09,0x00],
    "6":[0x0E,0x15,0x15,0x08,0x00],
    "7":[0x01,0x01,0x1D,0x03,0x00],
    "8":[0x0A,0x15,0x15,0x0A,0x00],
    "9":[0x02,0x15,0x15,0x0E,0x00],

    # Symbols
    "-":[0x04,0x04,0x04,0x00,0x00],
    " ":[0x00,0x00,0x00,0x00,0x00]
}

# ============================================================
# 5×5 FONT FOR LINE BULLETS
# ============================================================

FONT_5x5 = {
    "7": [
        0b11111,
        0b00010,
        0b00100,
        0b00100,
        0b00100
    ],
    "G": [
        0b01110,
        0b10000,
        0b10110,
        0b10010,
        0b01110
    ]
}

# ============================================================
# DRAWING FUNCTIONS
# ============================================================

def draw_char_5x7(draw, x, y, char, color):
    bitmap = FONT_5x7.get(char, FONT_5x7[" "])

    # Trim trailing empty columns
    trimmed = list(bitmap)
    while trimmed and trimmed[-1] == 0x00:
        trimmed.pop()

    for col, byte in enumerate(trimmed):
        for row in range(7):
            if byte & (1 << row):
                draw_led_pixel(draw, x + col, y + row, color)

def draw_text_5x7(draw, x, y, text, color):
    cursor_x = x

    for char in text:
        bitmap = FONT_5x7.get(char, FONT_5x7[" "])

        # Trim trailing empty columns
        trimmed = list(bitmap)
        while trimmed and trimmed[-1] == 0x00:
            trimmed.pop()

        # Draw the character
        for col, byte in enumerate(trimmed):
            for row in range(7):
                if byte & (1 << row):
                    draw_led_pixel(draw, cursor_x + col, y + row, color)

        # Advance cursor by actual width + 1 pixel spacing
        cursor_x += len(trimmed) + 1

def draw_char_5x5(draw, x, y, char, color):
    bitmap = FONT_5x5.get(char.upper())
    if bitmap is None:
        return
    for row in range(5):
        for col in range(5):
            if bitmap[row] & (1 << (4 - col)):
                draw_led_pixel(draw, x + col, y + row, color)

# ============================================================
# CANVAS SETUP
# ============================================================

WIDTH = 128
HEIGHT = 24

img = Image.new(
    "RGB",
    (
        WIDTH * (PIXEL_SIZE + PIXEL_GAP),
        HEIGHT * (PIXEL_SIZE + PIXEL_GAP)
    ),
    "black"
)
draw = ImageDraw.Draw(img)

ROW_H = 11
BUFFER = 2

row1_y = 0
row2_y = ROW_H + BUFFER

# ============================================================
# ROW RENDERING
# ============================================================
def measure_text_5x7(text):
    width = 0
    for char in text:
        bitmap = FONT_5x7.get(char, FONT_5x7[" "])
        trimmed = list(bitmap)
        while trimmed and trimmed[-1] == 0x00:
            trimmed.pop()
        width += len(trimmed) + 1   # actual width + spacing
    return width

def draw_row(draw, y, line_color, line_num, direction, time_text):

    # Row index (1. for top, 2. for bottom)
    row_index = "1." if y == 0 else "2."

    # Circle position
    circle_x = 12
    radius = 5

    # Both rows were 1px too high → shift both DOWN 1px
    if y == 0:
        circle_y = y + ROW_H // 2 + 1   # top row stays perfect
    else:
        circle_y = y + ROW_H // 2 - 1   # bottom row → move UP 1 pixel

    # Draw circle
    for yy in range(-radius, radius+1):
        for xx in range(-radius, radius+1):
            if xx*xx + yy*yy <= (radius * radius) - 2:
                draw_led_pixel(draw, circle_x + xx, circle_y + yy, line_color)

    # Center 5×5 line number
    draw_char_5x5(draw, circle_x - 2, circle_y - 2, line_num, "black")

    # Align all text to circle center
    text_baseline = circle_y - 2

    # Row index
    draw_text_5x7(draw, 2, text_baseline, row_index, "white")

    # Direction text
    draw_text_5x7(draw, 22, text_baseline, direction, "white")

    # Time text
    time_width = measure_text_5x7(time_text)
    draw_text_5x7(draw, WIDTH - time_width - 2, text_baseline, time_text, "white")

# ============================================================
# DRAW TWO ROWS
# ============================================================

draw_row(
    draw,
    row1_y,
    line_color=(128, 0, 128),
    line_num="7",
    # direction="34  St-Hudson  Ya",
    # time_text="3min"
    direction="34  ST-HUDSON  YA",
    time_text="3MIN"
)

draw_row(
    draw,
    row2_y,
    line_color=(0, 180, 0),
    line_num="G",
    # direction="Church  Ave",
    # time_text="5min"
    direction="CHURCH  AVE",
    time_text="5MIN"
)

# ============================================================
# SHOW RESULT
# ============================================================

display(img)
```
